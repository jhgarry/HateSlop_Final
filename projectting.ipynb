```python
receptors = ['sweet', 'sour', 'spicy', 'oily', 'salty', 'bitter']
```

```python
def get_similarity(list1, list2):
    # Encode each list as a joined sentence
    embed1 = word_model.encode(" ".join(list1), convert_to_tensor=True)
    embed2 = word_model.encode(" ".join(list2), convert_to_tensor=True)
    similarity = util.cos_sim(embed1, embed2).item()
    return similarity
```

```python
word_model = SentenceTransformer('all-MiniLM-L6-v2')
# def most_similar_flavor(data, predefined):
#     data_counter = Counter(data)
#     predefined_docs = {word: word_vec(word) for word in predefined}

#     similarity_scores = defaultdict(float)

#     for word, freq in data_counter.items():
#         word_doc = word_vec(word)
#         for flavor, flavor_doc in predefined_docs.items():
#             similarity = word_doc.similarity(flavor_doc)
#             similarity_scores[flavor] += similarity * freq

#     # Choose the predefined flavor word with the highest total similarity
#     return max(similarity_scores, key=similarity_scores.get)

def most_similar_flavor(data, predefined):
    data_counter = Counter(word.lower() for word in data)
    flavor_embeddings = word_model.encode(predefined, convert_to_tensor=True)

    similarity_scores = defaultdict(float)

    for word, freq in data_counter.items():
        word_embedding = word_model.encode(word, convert_to_tensor=True)
        cosine_scores = util.cos_sim(word_embedding, flavor_embeddings)[0]
        for i, score in enumerate(cosine_scores):
            similarity_scores[predefined[i]] += score.item() * freq

    if not similarity_scores:
        return "unknown"
    return max(similarity_scores, key=similarity_scores.get)
```

```python
matched_results = {}

for idx1, row1 in df_flavors.iterrows():
    flavor_id = row1[0]
    flavor_words = row1[1].lstrip("[").rstrip("]").replace("'", "").split(", ")

    best_match = None
    best_score = -1

    for idx2, row2 in df_grapes.iterrows():
        receptor_id = row2[0]
        receptor_words = row2[1].lstrip("[").rstrip("]").replace("'", "").split(", ")

        score = get_similarity(flavor_words, receptor_words)

        if score > best_score:
            best_score = score
            best_match = receptor_id

    matched_results[flavor_id] = best_match
```

# Training & Fine-tuning

```python
from datasets import load_dataset, Dataset
```

```python
wines_dataset = Dataset.from_pandas(df_wines)
wines_dataset
wines_dataset_train = wines_dataset.train_test_split(test_size=0.1, seed=37)['train']
wines_dataset_test = wines_dataset.train_test_split(test_size=0.1, seed=37)['test']
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
# def tokenizing_w(example):
#   return tokenizer(example['Wine'], padding=True, truncation=True)

# def tokenizing_v(example):
#   return tokenizer(example['Variety'], padding=True, truncation=True)

# def tokenizing_n(example):
#   return tokenizer(example['Note'], padding=True, truncation=True)

def tokenizing(example):
  text = f"Wine: {example['Wine']} Variety: {example['Variety']} Note: {example['Note']}"
  return tokenizer(text, padding=True, truncation=True)
```

```python
def format_example(example):
    input_text = f"Wine: {example['Wine']}\nVariety: {example['Variety']}\nNote:"
    target_text = example['Note']
    return {
        "prompt": input_text,
        "completion": target_text,
    }
```

```python
def tokenize(example):
    full_text = example["prompt"] + " " + example["completion"]
    tokenized = tokenizer(
        full_text,
        padding="max_length",
        truncation=True,
        max_length=512,
    )
    prompt_len = len(tokenizer(example["prompt"])["input_ids"])
    labels = tokenized["input_ids"][:]
    labels[:prompt_len] = [-100] * prompt_len

    tokenized["labels"] = labels
    return tokenized
```

```python
# wines_dataset_train = wines_dataset_train.map(tokenizing_w, batched=True)
# wines_dataset_test = wines_dataset_test.map(tokenizing_w, batched=True)
# wines_dataset_train = wines_dataset_train.map(tokenizing_v, batched=True)
# wines_dataset_test = wines_dataset_test.map(tokenizing_v, batched=True)
# wines_dataset_train = wines_dataset_train.map(tokenizing_n, batched=True)
# wines_dataset_test = wines_dataset_test.map(tokenizing_n, batched=True)

wines_dataset_train = wines_dataset_train.map(tokenizing, batched=True)
wines_dataset_test = wines_dataset_test.map(tokenizing, batched=True)
```

```python
formatted_train = wines_dataset_train.map(format_example)
formatted_test = wines_dataset_train.map(format_example)
```

```python
formatted_train = formatted_train.map(tokenize, remove_columns=formatted_train.column_names)
formatted_test = formatted_test.map(tokenize, remove_columns=formatted_test.column_names)
```

```python
training_args = TrainingArguments(
    output_dir='/content/drive/MyDrive/project/git/HateSlop_Final',
    num_train_epochs=1,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    eval_strategy='epoch',
    learning_rate=2e-5,
    remove_unused_columns=False,
    push_to_hub=True
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=formatted_train,
    eval_dataset=formatted_test,
    tokenizer=tokenizer,
)
```

```python
trainer.train()
```
